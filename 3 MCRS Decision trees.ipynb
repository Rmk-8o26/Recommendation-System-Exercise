## Step 1 : Preperation of Data

```python
import pandas as pd
df=pd.DataFrame([[3.0,2.0,1.0],[1.0,1.0,4.0],[4.0,2.0,1.0],[1.0,2.0,3.0],[2.0,3.0,1.0]])
X=df.iloc[:,0:2]
y=df.iloc[:,2]
#X
#y
```

## Step 2 : Dec Tree for Predicting $i_1$

```python
X=df.iloc[:,[1,2]]
y=df.iloc[:,0]
from sklearn import tree
import matplotlib.pyplot as plt
clf = tree.DecisionTreeClassifier()
model1=clf.fit(X,y)
fig=plt.figure(figsize=(15,5))
_=tree.plot_tree(clf,filled=True)
```

## Step 3 : Dec Tree for Predicting $i_2$

```python
X=df.iloc[:,[0,2]]
y=df.iloc[:,1]
model2=clf.fit(X,y)
fig=plt.figure(figsize=(15,5))
_=tree.plot_tree(clf,filled=True)
```

## Step-4: Dec Tree for Predicting $i_3$

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
model3=clf.fit(X,y)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(15,5))
_=tree.plot_tree(clf,filled=True)
```

## Textual Representation of third decision tree

```python
text_representation=tree.export_text(clf)
print(text_representation)
```

## Out of sample prediction of rating for item-1 given the values of other features are 1 and 2

```python
t2=np.array([[3.0,3.0],[1.0,2.0]])
print(model3.predict(t2))
```
