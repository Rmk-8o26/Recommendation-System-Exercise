## Problem : Consider the data given below related to 20 users and 5 items. Estimate the missing ratings usingNaive Bayesian Classifier


### Here we need to develop three models. Model 1 willbe used for predicting the rating of 7th user for the 2nd item,Model 2 is meant for predicting the rating of 8th user for 3rd item and Model 3 is meant for predicting the rating of 19th user for 5th item

## Step 1: Importing the given data

```python
import pandas as pd
df=pd.read_csv("C:/Users/Mohan/ds/nb-rec-data.csv")
df
```

### Slicing the rows and columns with no missing values

```python
df_full=df.dropna()
df_full
```

## Step-2: Development of Model-1

```python
from sklearn.naive_bayes import CategoricalNB
```

### Splitting of target and feature variables

```python
X = df_full.iloc[:,[0,2,3,4]]
y = df_full.iloc[:,1]
```

### Fit the model

```python
model1 = CategoricalNB()
model1.fit(X,y)
```

### Prediction using Model-1 - rating of 7th user for item 2

```python
import numpy as np
test=np.array([[1,1,2,1]])
model1.predict(test)
```
